```python
import pandas as pd
```

```python
df1 = pd.read_csv('discogs_tracks_2007_with_lyrics.csv')
df2 = pd.read_csv('musicbrainz_tracks_2007_offset0_with_lyrics.csv')
```

```python
# Can concat no matching columns (what we've set up below) or overlapping
# columns and returning everything (like a FULL JOIN)
#df3 = pd.concat([df1, df2], sort=False)
#df3.head()
```

```python
df1.drop(columns='track_position', inplace=True)

df1.rename(columns={"album_title": "album", "track_title": "title", "release_id": "id"}, inplace=True)

new_order = ['title', 'duration', 'album', 'year', 'artist', 
            'id', 'country', 'genre', 'style', 'source', 'lyrics']
df1 = df1[new_order]

def convert_to_seconds(value):
    try:
        minutes, seconds = map(int, str(value).split(':'))
        return minutes * 60 + seconds
    except:
        return None  # or np.nan

df1['duration'] = df1['duration'].apply(convert_to_seconds)

df1.insert(3, "release_date", None)
```

```python
df1.head()
```

```python
df2.rename(columns={"track_length_ms": "duration", "MBID": "id"}, inplace=True)
df2['duration'] = (df2['duration'] / 1000).round(2)
df2.insert(4, "year", None)
df2.insert(7, "country", None)
df2.insert(8, "genre", None)
df2.insert(9, "style", None)
```

```python
df2.head()
```

```python
df3 = pd.concat([df1, df2])
df3.head()
```
